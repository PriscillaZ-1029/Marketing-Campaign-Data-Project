```python
t=df2.groupby(by='education').mean()["age"]
```

```python
label=df2["education"].unique().tolist()
label
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        # 
        return '{p:.2f}% ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
plt.figure(dpi=150,figsize=(8,5))
plt.pie(t,labels=label
       ,autopct=make_autopct(t.tolist()),shadow=True,colors=["red","green","#65a479","cyan","lavender"],labeldistance=1.2
       ,startangle=90,explode=(0,0,0.3,0))
plt.title('The relationship between age and client education level')
plt.legend(loc="best",fontsize="small"
         ,title_fontsize=6,markerfirst=True, #Place the legend on the left side
           bbox_to_anchor=(0.9, 0., 0.5, 0.3),title="Victim Sex") #Position the legend
plt.show()
```

```python
label2=df2["job"].unique().tolist()
label2
```

```python
t2=df2.groupby(by='job').mean()["age"]
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
ax=sns.barplot(x=label2, y=t2)
ax.set_xticklabels(labels=label2,rotation=78,fontsize=15)
plt.xlabel('Job title',fontsize=18)
plt.xlabel('Age',fontsize=18)
plt.title('Average age of clients based on job')
plt.show()
plt.tight_layout()
```
